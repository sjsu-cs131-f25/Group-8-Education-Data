```python
from google.colab import drive
drive.mount('/content/drive')

!ls /content/drive/MyDrive/Group8_EducationData

data_path = "/content/drive/MyDrive/Group8_EducationData/student_performance.csv"
student_df = spark.read.csv(data_path, header=True, inferSchema=True)
student_df.printSchema()
student_df.show(5)
```

```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window
```

```python
# Cleaning the dataset i.e. numbers are converted into names for categories.

df = (student_df.withColumns({
    "Resources": F.col("Resources").cast("string"),
    "Extracurricular": F.col("Extracurricular").cast("string"),
    "Gender": F.col("Gender").cast("string"),
    "LearningStyle": F.col("LearningStyle").cast("string"),
    "Discussions": F.col("Discussions").cast("string"),
    "EduTech": F.col("EduTech").cast("string"),
    "FinalGrade": F.col("FinalGrade").cast("string")
}))

df = (df.withColumns({
    "Resources": F.expr("""CASE WHEN Resources == '1' THEN 'Yes'
      ELSE 'No' END AS Resources"""),
    "Extracurricular": F.expr("""CASE WHEN Extracurricular == '1' THEN 'Yes'
      ELSE 'No' END AS Extracurricular"""),
    "Gender": F.expr("""CASE WHEN Gender == '1' THEN 'Female'
      ELSE 'Male' END AS Gender"""),
    "LearningStyle": F.expr("""CASE WHEN LearningStyle == '0' THEN 'Visual'
      WHEN LearningStyle == '1' THEN 'Auditory'
      WHEN LearningStyle == '2' THEN 'Kinesthetic'
      WHEN LearningStyle == '3' THEN 'Reading/Writing' END AS LearningStyle"""),
    "Discussions": F.expr("""CASE WHEN Discussions == '1' THEN 'Yes'
      ELSE 'No' END AS Discussions"""),
    "EduTech": F.expr("""CASE WHEN EduTech == '1' THEN 'Yes'
      ELSE 'No' END AS EduTech"""),
    "FinalGrade": F.expr("""CASE WHEN FinalGrade == '0' THEN 'A'
      WHEN FinalGrade == '1' THEN 'B'
      WHEN FinalGrade == '2' THEN 'C'
      WHEN FinalGrade == '3' THEN 'D' END AS FinalGrade""")
}))
```

```python
df.show()
df.printSchema()
```

```python
# Generate data of students who use one of the four learning styles.

learning_df = (df
             .groupBy("LearningStyle")
             .agg(F.count("LearningStyle").alias("Students"))
             .orderBy(F.desc("Students")))

learning_df.show()
```

```python
# Generate the number of students who frequently use the internet for their studies.

edutech_df = (df
            .groupBy("EduTech")
            .agg(F.count("EduTech").alias("Students"))
            .orderBy(F.desc("Students")))

edutech_df.show()
```

```python
# Generate top 10 weekly study hours amongst students.

window = Window.orderBy(F.desc("Students"))

temp = (df
      .groupBy("StudyHours")
      .agg(F.count("StudyHours").alias("Students"))
      .orderBy(F.desc("Students")))

temp2 = temp.withColumn("rank", F.row_number().over(window))

weekly_studyhours_df = (temp2.select("rank","StudyHours","Students"))

weekly_studyhours_df.show(10)
```

```python
# Generate a skinny table containing the columns attendance, extracurricular, and exam scores.

skinny_df = (df
             .select("Attendance", "Extracurricular", "ExamScore")
             .orderBy(F.desc("ExamScore")))

skinny_df.show()
```

```python
# With the skinny table, only generate data of students who scored 60-69% on their exam.

examscore60s_df = (skinny_df
                   .filter(skinny_df.ExamScore >= 60)
                   .filter(skinny_df.ExamScore < 70)
                   .orderBy(F.asc("ExamScore")))

examscore60s_df.show()
```

```python
# With the skinny table, only generate data of students who scored 90% or higher.

examscore90s_df = (skinny_df
                   .filter(skinny_df.ExamScore >= 90)
                   .orderBy(F.asc("ExamScore")))

examscore90s_df.show()
```

```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window

# Quick overview
print("Total records:", df.count())
print("Total columns:", len(df.columns))

# Average and summary stats grouped by Gender
summary_df = df.groupBy("Gender").agg(
    F.count("*").alias("student_count"),
    F.avg("StudyHours").alias("avg_study_hours"),
    F.avg("Attendance").alias("avg_attendance"),
    F.avg("ExamScore").alias("avg_exam_score"),
    F.avg("FinalGrade").alias("avg_final_grade")
)
summary_df.show()

# Top 3 students by ExamScore within each Gender
windowSpec = Window.partitionBy("Gender").orderBy(F.desc("ExamScore"))
ranked_df = df.withColumn("rank", F.rank().over(windowSpec))
ranked_df.filter(F.col("rank") <= 3).show()
```

```python
from pyspark.sql import functions as F

# Table 1: Study Habits Summary
study_habits_df = df.groupBy("Gender").agg(
    F.avg("StudyHours").alias("avg_study_hours"),
    F.avg("Attendance").alias("avg_attendance"),
    F.avg("Motivation").alias("avg_motivation"),
    F.avg("ExamScore").alias("avg_exam_score")
)

study_habits_df.show()
```

```python
# Bucket StressLevel into Low/Medium/High categories
stress_df = df.withColumn(
    "StressCategory",
    F.when(F.col("StressLevel") <= 3, "Low")
     .when((F.col("StressLevel") > 3) & (F.col("StressLevel") <= 7), "Medium")
     .otherwise("High")
)

# Group by stress category
stress_summary_df = stress_df.groupBy("StressCategory").agg(
    F.count("*").alias("student_count"),
    F.avg("ExamScore").alias("avg_exam_score")
)

stress_summary_df.show()
```
